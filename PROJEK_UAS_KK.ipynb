```python
data = pd.read_csv('healthcare-dataset-stroke-data.csv', sep = ',') 
data
```

```python
#menghitung total data yang kosong di dataframe
#persen
print(data.isnull().sum()/len(data)*100)
print("="*20)
print(data.isna().sum())
```

```python
# data = data.dropna(how='any')           # assign back
data.dropna(how='any', inplace=True) 
# data.dropna()
data
```

```python
#data unkonwn == data Nan
data[data["smoking_status"] == 'Unknown']
```

```python
# data=data.drop(data["smoking_status"] == 'Unknown')
data = data[data['smoking_status'] != 'Unknown']
data
```

```python
data.isna().sum()
```

```python
# mengubah data berupa kategori menjadi dalam bentuk angka
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data['gender'] = le.fit_transform(data['gender'])
data['ever_married'] = le.fit_transform(data['ever_married'])
data['work_type'] = le.fit_transform(data['work_type'])
data['Residence_type'] = le.fit_transform(data['Residence_type'])
data['smoking_status'] = le.fit_transform(data['smoking_status'])
data
```

```python
# Memisahkan kolom stroke dengan kolom lainnya
x=data.drop(['stroke'], axis=1)
y=data['stroke']

# Membagi data menjadi data train dan data test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state= 200)

svm = SVC()

# Membuat model svm berdasarkan data train yang diberikan
svm.fit(x_train,y_train)
```

```python
# Membuat prediksi
predictions = svm.predict(x_test)
# Mendapatkan score akurasi prediksi
acc_score = accuracy_score(y_test, predictions)

print("Akurasi Model SVM  : {}".format(format(acc_score,'.3f')), '\n')
print("Confusion matriks : \n", confusion_matrix(y_test, predictions),"\n")
```
